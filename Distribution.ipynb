#### outcome_2

```python
sns.distplot(x,
            bins=20,
            hist=True,
            kde=True,
            hist_kws={'color':'g', 'histtype':'bar', 'alpha':0.4},
            kde_kws={'color':'b', 'linestyle':'-', 'linewidth':1, 'alpha':0.4});
```

### example_1

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.cyl.value_counts()
```

```python
plt.figure(figsize=(14,10), dpi=65)

sns.kdeplot(df.loc[df.cyl == 4, 'cty'], shade=True, color='green', alpha=0.5, label='Cyl = 4')
sns.kdeplot(df.loc[df.cyl == 5, 'cty'], shade=True, color='deeppink', alpha=0.5, label='Cyl = 5')
sns.kdeplot(df.loc[df.cyl == 6, 'cty'], shade=True, color='blue', alpha=0.5, label='Cyl = 6')
sns.kdeplot(df.loc[df.cyl == 8, 'cty'], shade=True, color='orange', alpha=0.5, label='Cyl = 8')

plt.title('Density Plot Of City Mileage by n_Cylinders', fontsize=22)
plt.xlabel('City Mileage', fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=13);
```

### example_2

```python
clas = list(set(df['class'].values))
clas
```

```python
plt.figure(figsize=(13,8), dpi=65)

for i in clas:
    sns.distplot(df.loc[df['class'] == i, 'cty'].values, label=i)
    
plt.title('Density Plot Of City Mileage by n_Cylinders', fontsize=22)
plt.ylim(0,0.7)
plt.xlim(5,30)
plt.legend(fontsize=15);
```

```python
class_data = ['minivan', 'suv', 'compact']

plt.figure(figsize=(13,8), dpi=65)

sns.distplot(df.loc[df['class'] == 'minivan', 'cty'].values, color='green', label='minivan',
             hist_kws={'alpha':.7},
             kde_kws={'linewidth':2, 'alpha':.7})

sns.distplot(df.loc[df['class'] == 'suv', 'cty'].values, color='orange', label='suv',
             hist_kws={'alpha':.7},
             kde_kws={'linewidth':2, 'alpha':.7})

sns.distplot(df.loc[df['class'] == 'compact', 'cty'].values, color='dodgerblue', label='compact',
             hist_kws={'alpha':.7},
             kde_kws={'linewidth':2, 'alpha':.7})
    
plt.title('Density Plot Of City Mileage by n_Cylinders', fontsize=22)
plt.xlabel('City Mileage', fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=15);
```
