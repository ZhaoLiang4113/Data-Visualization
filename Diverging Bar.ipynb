### Drverging Bars

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.rand(10)
x = x-x.mean()
x.sort()
#sorted(x,reverse=True)
```

```python
colors = []
for i in x:
    if i > 0:
        colors.append('r')
    else:
        colors.append('b')
```

```python
plt.figure(figsize=(8,4))
plt.hlines(y=range(10),xmin=0,xmax=x,
          linewidth=3,
          color=colors,
          alpha=0.6);
```

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
df.head()
```

```python
df.shape
```

```python
sum(df.cars==df.carname)
```

```python
x = df['mpg']
df['mpg_z'] = (x - x.mean())/x.std()
df.sort_values('mpg_z', inplace=True)
df.reset_index(drop=True, inplace=True)
df['colors'] = ['g' if i > 0 else 'r' for i in df['mpg_z']]
```

```python
plt.figure(figsize=(14,10), dpi=65)
plt.hlines(y=df.cars, xmin=0, xmax=df.mpg_z,
          color=df.colors,
          alpha=0.5,
          linewidth=5)
plt.ylabel("$models$", fontsize=18)
plt.xlabel("$mileages$", fontsize=18)
plt.yticks(fontsize=12)
plt.title('Divergings Bars of Car Mileages', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5);
```
