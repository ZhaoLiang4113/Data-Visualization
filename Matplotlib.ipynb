```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

data = randn(50).cumsum()

ax1.hist(randn(100), bins=20, color='g', alpha=0.5, label='bar')
ax2.scatter(np.arange(100), np.arange(100) + 3 * randn(100), color='b', alpha=0.7, label='point')
ax3.plot(data, color='g', alpha=0.7, linestyle='-', marker='o', label='line')
ax4.plot(data, color='k', alpha=1, linestyle='-', label='deault')
ax4.plot(data, color='r', alpha=1, linestyle='-', drawstyle='steps-post', label='step-post')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best');
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0);
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1, 1, 1)

colors = ['r', 'g', 'b']
labels = ['One', 'Two', 'Three']
linestyles = ['-', '--', '-']

for i in range(3):
    ax.plot(randn(100).cumsum(), color=colors[i], alpha=0.5, linestyle=linestyles[i], linewidth=1.5, label=labels[i])

ticks = ax.set_xticks([0, 25, 50, 75, 100])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], fontsize=12, rotation=0)

ax.set_title('My first matplotlib plot', fontsize=18)
ax.set_xlabel('Stages', fontsize=15)
ax.legend(loc='best');
# ax.text(x, y, 'Hello world', family='monspace', fontsize=10);
```

```python
from datetime import datetime
import pandas as pd

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='b-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label, in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 50),
                xytext=(date, spx.asof(date) + 20),
                arrowprops=dict(facecolor='black'),
                horizontalalignment='right', verticalalignment='baseline')
    # ('top', 'bottom', 'center', 'baseline', 'center_baseline')

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in 2008-2009 financila crisis', fontsize=22); 
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

plt.savefig('figpath.png',facecolor='gray', dpi=400, bbox_inches='tight');
```
